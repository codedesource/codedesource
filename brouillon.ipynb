```python
url = "https://bnpe.eaufrance.fr/acces-donnees/codeDepartement/66/annee/2020/" 

user_agent = "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.149 Safari/537.36"
headers = {"User-Agent": user_agent}
```

```python
from bs4 import BeautifulSoup
import requests

# Make a request to the webpage you want to scrape
response = requests.get(url, headers=headers)
# Create a BeautifulSoup object to parse the HTML
soup = BeautifulSoup(response.text, "html.parser")
# Use CSS selectors to extract specific elements from the HTML
# For example, let's extract all the links (anchor tags) from the page
myselector="#table-synthesegeographique *"
print(soup)
content = soup.select(myselector)
# Print the text and href attributes of each link
for link in content:
    print("Text:", link.text)
```
