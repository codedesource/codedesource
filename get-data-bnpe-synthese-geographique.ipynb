# Récupérer les données BNPE

## La fonction

```python
import requests
from jsonpath_ng import parse # https://github.com/h2non/jsonpath-ng

def récupérer_données_bnpe(url: str, jsonpath_search_pattern: str) -> (list|dict):
    à_retourner = []
    json_brut = requests.get(url).json()
    if json_brut is None:
        return à_retourner
    données_dintérêt = [match.value for match in parse(jsonpath_search_pattern).find(json_brut)]
    if données_dintérêt is None:
        return à_retourner
    return données_dintérêt
```

## Données « Synthèse géographique »

Département 66, année 2020

```python
annee = 2020
code_dep = 66
url = f'https://bnpe.eaufrance.fr/Bnpe-Diffusion/synthese/synthese_geographique?&code_dep={code_dep}&annee={annee}&ecrasant'
pattern = '$.features[*].properties'
synthese_geographique = récupérer_données_bnpe(url, pattern)
```
