# Présentation :Ce notebook permet de transférer un CSV dans une table Big Query

```python
# Import des librairies
from google.cloud import bigquery
from google.cloud import storage
import os
```

```python
# Initialisation des clients BigQuery et GCS
bigquery_client = bigquery.Client()
storage_client = storage.Client()
```

```python
# Nom de la table BigQuery
table_id = 'code-de-source.donnees_code_de_source.consommationEau'
```

```python
# Nom du bucket GCS
bucket_name = 'code_de_source_lake'
```

```python
# Récupérer le bucket
bucket = storage_client.bucket(bucket_name)
```

```python
# Liste pour stocker les noms des fichiers
file_names = []
```

```python
# Parcourir les objets (fichiers) dans le bucket
for blob in bucket.list_blobs():
    file_names.append(blob.name)
```

```python
# Afficher les noms des fichiers
for gcs_file_path in file_names:
    # Générer l'URI du fichier dans GCS
    gcs_uri = f"gs://{bucket_name}/{gcs_file_path}"
    
    # Charger les données dans la table BigQuery
    table = bigquery_client.get_table(table_id)
    job_config = bigquery.LoadJobConfig(skip_leading_rows=1, source_format=bigquery.SourceFormat.CSV)
    job = bigquery_client.load_table_from_uri(gcs_uri, table, job_config=job_config)

    job.result()  # Attendre que le chargement soit terminé
    print(f"Le fichier {gcs_file_path} a été envoyé dans la table {table_id}")
    
```
